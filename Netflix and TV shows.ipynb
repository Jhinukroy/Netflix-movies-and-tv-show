## **Problem statement:** Cleaning and Preprocessing from Raw data. 

### Data Preprocessing:
 **Uploading the dataset using pandas , treat missing values and duplicates ,standardize text values, Convert date format to a consistent type, rename column headers to be clean and uniform, check & fix data type**

```python
# importing necessary libraries like pandas and numpy
import numpy as np
import pandas as pd
```

```python
# For reading the data set in python
Data= pd.read_csv(r"C:\Users\MEW\OneDrive\Documents\elevate Labs dataset\Netflix and Show.csv",header=0)
```

```python
Data.head()
```

We can see some Columnns are not required for analyzing the dataset.

```python
Data.isna().sum()  # for check sum of missing values 
```

As you can see there is missing values in column named director,cast,country,date_added,In rating 4 missing values are there. Duration has 3 missing values. 

```python
del Data['show_id']   # delete unnecessary column
```

```python
Data.head() # displaying first 5 rows
```

```python
# Filling NaN in director, cast, and country with 'Unknown'
Data[['director', 'cast', 'country']] = Data[['director', 'cast', 'country']].fillna('Unknown')
```

```python
Data.head()
```

```python
# Impute missing values in 'date_added' with the mode
Data['date_added'].fillna(Data['date_added'].mode()[0], inplace=True)

# Impute missing values in 'rating' with the mode
Data['rating'].fillna(Data['rating'].mode()[0], inplace=True)

# Impute missing values in 'duration' with the mode
Data['duration'].fillna(Data['duration'].mode()[0], inplace=True)
```

We will use mode() because categorical values are missing.

```python
Data.isna().sum()
```

```python
df = Data.drop_duplicates()
```

```python
df.head() #with no duplicate 
```

```python
# Columns with text values that might need standardization
text_columns = ['type', 'title', 'director', 'cast', 'country', 'rating', 'duration', 'listed_in', 'description']

for col in text_columns:
    if df[col].dtype == 'object':  # Only apply to string columns
        df[col] = df[col].str.lower()
```

```python
df.sample(10)
```

```python
 # Trying converting 'date_added' to datetime with 'format='mixed''
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed', dayfirst=False, errors='coerce')
```

```python
df.head(3)
```

**here you can see the date format has been changed**

```python
# Rename columns to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')
```
